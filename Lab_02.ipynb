How many different occupations are in this dataset?

````python
# Displaying different occupations are in this dataset?
total = len(users["occupation"].unique())
print(f"There are {total} different occupations in the Data Set")
````

What is the most occuring occupation for each gender seperatly?

````python
# Checking the most occuring occupation for each gender seperatly
Males = users[users["gender"] == "M"]
Females = users[users["gender"] == "F"]
Mmost = Males["occupation"].max()
Fmost = Females["occupation"].max()
print(f"The most occuring occupation for Males is {Mmost}")
print(f"The most occuring occupation for Males is {Fmost}")
````

What is the mean age of users?

````python
# Calculating the Mean Age of users
usermean = round(users["age"].mean(), 2)
print(f"Mean age for users is {usermean}")
````

````python
# Displaying the users who have age more than 40
Ndf = users[users["age"]>40]
Ndf
````

````python
# Finding max and min value from each numeric column
mini = users["age"].min()
maxi = users["age"].max()
print(f"Minimum Value in Age Column is {mini} and Maximum is {maxi}")
````
